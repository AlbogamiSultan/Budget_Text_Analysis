```python
from statsmodels.stats import weightstats as stests
```

```python
ztest ,pval1 = stests.ztest(CCE.sentiment, x2=CCE20.sentiment, value=0,alternative='two-sided')
print(float(pval1))
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
