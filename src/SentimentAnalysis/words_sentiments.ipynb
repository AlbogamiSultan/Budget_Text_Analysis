```python
#Function to display a wordcloud
def budget_word_cloud(data,sentiment):
    budget_sentiment = data[data['sentiment'] == sentiment]
    words = " ".join(budget_sentiment['word'])
                 
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(words)
                 
    plt.figure(1,figsize=(12, 12))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    return (words)
```

```python
#Plot a wordcloud for negative sentiments
CCE_negative_words = budget_word_cloud(CCE,"Negative")
```

```python
#Plot a wordcloud for negative sentiments
CCE_positive_words = budget_word_cloud(CCE,"Positive")
```
